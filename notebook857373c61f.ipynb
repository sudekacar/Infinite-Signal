```python
import plotly.express as px
import pandas as pd
import os

# 1. ADIM: Kaggle'daki gerçek yolu listele ve bul
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.csv'):
            actual_path = os.path.join(dirname, filename)
            print(f"Buldum! Gerçek yolun bu: {actual_path}")

# 2. ADIM: Bulduğumuz yolu kullanarak veriyi oku
df = pd.read_csv(actual_path)

# 3. ADIM: Görselleştirme (Severity ve Population kolonlarının varlığından emin ol)
# Eğer kolon isimlerin farklıysa (büyük harf vs) burayı ona göre güncelle
fig = px.scatter_mapbox(df, lat="latitude", lon="longitude", color="severity", 
                        size="population", hover_name="zone", zoom=5)

fig.update_layout(mapbox_style="open-street-map", margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
